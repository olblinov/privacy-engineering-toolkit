```python
import json, yaml
secrets_path = 'secrets.json'
with open(secrets_path, 'r') as file: 
    secrets = json.load(file)
```

```python
config = """
schema_version: "1.0"
name: classify-default
models:
  - classify:
      data_source:
        - gretel_83885c601bc147f0bf7bda6752342ba2_orders.csv
      ref_data: {}
      use_nlp: false
      return_matches: false
      labels:
        - person_name
        - credit_card_number
        - phone_number
        - us_social_security_number
        - email_address
        - location
notifications: null
label_predictors: null

"""
```

```python
from gretel_client import Gretel
import pandas as pd
from gretel_client.helpers import poll
gretel = Gretel(project_name=secrets['project_name'], api_key=secrets['api_key'])
project = gretel.get_project()
```

```python
classify = project.create_model_obj(
    model_config=yaml.safe_load(config), data_source="../../assets/datasets/orders.csv"
)
```

```python
classify.submit_cloud()
```

```python
import requests
import gzip
import json

def download_and_load_json(url):
    # Download the gzip file from the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Decompress the gzip data
        decompressed_data = gzip.decompress(response.content)
        
        # Load JSON data from the decompressed content
        json_data = json.loads(decompressed_data)
        
        return json_data
    else:
        # If request failed, print an error message
        print(f"Failed to download data. Status code: {response.status_code}")
        return None

json_data = download_and_load_json(classify.get_artifact_link("report_json"))
```
