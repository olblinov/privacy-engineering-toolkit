# Original Dataset

Non-sensitive data: Zip Code, Age, Nationality

Sensitive data: Condition

```python
data = pd.read_csv('../../assets/datasets/k_dataset.csv')

data
```

# K-anonymity

## Creating a k-anonymous dataset

```python
data.loc[:4, 'Zip Code'] = '130*'
data.loc[:4, 'Age'] = '<30'
data.loc[:4, 'Nationality'] = '*'

data.loc[5:8, 'Zip Code'] = '1485*'
data.loc[5:8, 'Age'] = '>40'
data.loc[5:8, 'Nationality'] = '*'

data.loc[9:12, 'Zip Code'] = '130*'
data.loc[9:12, 'Age'] = '3*'
data.loc[9:12, 'Nationality'] = '*'

data
```

## Attack on k-anonymous dataset

```python
attack_dataset = data = pd.DataFrame({
    "Name":["Bob"],
    "Age":["31"],
    "Zip Code":["13053"],
})

attack_dataset
```

Bob can be one of rows 9-12. All of these rows have the condition "Cancer". Ergo, Bob has cancer.
